### Renomeando colunas do dataset para PT-BR

```python
dados.rename(columns = {
    'Country_code': 'ID', 
    'Country': 'Pais', 
    'Total Library Size': 'Tamanho do catalogo', 
    'No. of TV Shows': 'Numero de series',
    'No. of Movies': 'Numero de filmes',
    'Cost Per Month - Basic ($)': 'Valor assinatura basica',
    'Cost Per Month - Standard ($)': 'Valor assinatura padrao',
    'Cost Per Month - Premium ($)': 'Valor assinatura premium'}, inplace = True)
dados
```

### Verificando distribui√ß√£o de frequ√™ncia com o Boxplot

```python
import seaborn as sns
ax = sns.boxplot(data=dados['Tamanho do catalogo'], orient = 'h', width = 0.3)
ax.figure.set_size_inches(12,6)
ax.set_title('Tamanho do cat√°logo', fontsize = 16)
ax
```

### O n√∫mero de t√≠tulos est√° relacionado com o valor da assinatura?

```python
dados.corr()
```

```python
ax = sns.lmplot(x='Valor assinatura basica', y = 'Tamanho do catalogo', data = dados)
ax.fig.set_size_inches(12,6)
ax.fig.suptitle('Valor da assinatura x Assinatura b√°sica', fontsize = 16)
ax.set_xlabels('Valor da assinatura b√°sica', fontsize = 14)
ax.set_ylabels('Tamanho do cat√°logo', fontsize = 14)
ax
```

```python
ax = sns.lmplot(x='Valor assinatura padrao', y = 'Tamanho do catalogo', data = dados)
ax.fig.set_size_inches(12,6)
ax.fig.suptitle('Valor da assinatura x Assinatura padr√£o', fontsize = 16)
ax.set_xlabels('Valor da assinatura padr√£o', fontsize = 14)
ax.set_ylabels('Tamanho do cat√°logo', fontsize = 14)
ax
```

```python
ax = sns.lmplot(x='Valor assinatura premium', y = 'Tamanho do catalogo', data = dados)
ax.fig.set_size_inches(12,6)
ax.fig.suptitle('Valor da assinatura x Assinatura premium', fontsize = 16)
ax.set_xlabels('Valor da assinatura premium', fontsize = 14)
ax.set_ylabels('Tamanho do cat√°logo', fontsize = 14)
ax
```

Ou seja, em nenhum modelo de assinatura, existe uma correla√ß√£o com o tamanho do cat√°logo dos diversos pa√≠ses. Em todos os casos, a correla√ß√£o foi inclusive levemente negativa.
